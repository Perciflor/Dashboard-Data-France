```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
# https://www.journaldunet.com/management/ville/index/villes?page=1
url = "https://www.journaldunet.com/management/ville/index/villes?page=1"
```

```python
req = requests.get(url)
```

```python
# je récupère le contenu
contenu = req.content
```

```python
# Je réécupère la soupe
soup = bs(contenu, "html.parser")
```

```python
groupelistes = soup.findAll('ul', class_ = "bloc size1of2 odSquareList")
```

```python
for i in range(1, 709):
    url = "https://www.journaldunet.com/management/ville/index/villes?page=" + str(i)
    soup = bs(requests.get(url).content, "html.parser")
    tousLesLiens = soup.findAll('a')
    for lien in tousLesLiens:
        if '/ville-' in lien['href']:
            print(lien['href'])
```
